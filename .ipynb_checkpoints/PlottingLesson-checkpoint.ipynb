# Plotting Lesson

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
distFile = os.path.join('data', 'distance_data_headers.csv')
```

```python
distances = np.genfromtxt(fname=distFile, delimiter=',', dtype='unicode')
#print(distances)
```

```python
headers = distances[0]
data = distances[1:].astype(np.float)   # If you do not recast the data as float, you will not get a plot!
#print(headers)
#print(data)
```

```python
plt.figure()   # Initializing a new figure
```

```python
plt.plot(data[:,1])
```

```python
sample = headers[1]

plt.figure()

plt.xlabel('Simulation Frame')
plt.ylabel(r'Distance [$\mathrm{\AA}$]')   # This is how to write LaTeX formatting: r

Fig1 = plt.plot(data[:,1], label=sample)

plt.legend()

plt.savefig(F'{sample}.png')
```

```python
plt.figure()

plt.xlabel('Simulation Frame')
plt.ylabel(r'Distance [$\mathrm{\AA}$]') 

plt.plot(data[:,0], data[:,1], label=headers[1])
plt.plot(data[:,0], data[:,2], label=headers[2])

plt.legend()
```

```python
plt.figure()

plt.xlabel('Simulation Frame')
plt.ylabel(r'Distance [$\mathrm{\AA}$]')

for column in range(1, len(headers)):
    plt.plot(data[:,0], data[:,column], label=headers[column])

plt.legend()
plt.savefig('allSamples.png', dpi=300)   # Increased resolution!
```

```python
for column in range(1, len(headers)):
    plt.figure()
    
    plt.plot(data[:,0], data[:,column], label=headers[column])
    
    plt.xlabel('Simulation Frame')
    plt.ylabel(r'Distance [$\mathrm{\AA}$]') 
    
    plt.legend()
```

```python
print(data[0:10:2, :])
```

```python
plt.figure()

plt.xlabel('Simulation Frame')
plt.ylabel(r'Distance [$\mathrm{\AA}$]')

for column in range(1, len(headers)):
    plt.plot(data[0::100,0], data[0::100,column], label=headers[column])   # Plot every 100 steps "0::100"

plt.legend()
plt.savefig('allSamples.png', dpi=300)
```

```python
plt.figure()

plt.xlabel('Simulation Frame')
plt.ylabel(r'Distance [$\mathrm{\AA}$]')

for column in range(1, len(headers)):
    plt.plot(data[0::100,0], data[0::100,column], '-.o', label=headers[column])   # Different plotting style.

plt.legend()
plt.savefig('allSamples.png', dpi=300)
```
